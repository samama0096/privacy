# Applying sobel operator without any built-in functon:


```python
import numpy as np
from PIL import Image

def grayscale(image):
    return np.dot(image[..., :3], [0.2989, 0.5870, 0.1140]).astype(np.uint8)

def sobel_operator(image):
    gray_image = grayscale(image)

    sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

    gradient_x = np.zeros_like(gray_image, dtype=float)
    gradient_y = np.zeros_like(gray_image, dtype=float)

    for y in range(1, gray_image.shape[0] - 1):
        for x in range(1, gray_image.shape[1] - 1):
            gradient_x[y, x] = np.sum(sobel_x * gray_image[y-1:y+2, x-1:x+2])
            gradient_y[y, x] = np.sum(sobel_y * gray_image[y-1:y+2, x-1:x+2])

    gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2).astype(np.uint8)

    return gradient_magnitude

# Load an image using PIL

#change path here:
image = Image.open('/content/drive/MyDrive/OIP.jpeg')
image_array = np.array(image)
sobel_result = sobel_operator(image_array)

#print(sobel_result)
```

# Applying prewitt operator without any built-in functon:


```python

def prewitt_operator(image):
    gray_image = grayscale(image)

    prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
    prewitt_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])

    gradient_x = np.zeros_like(gray_image, dtype=float)
    gradient_y = np.zeros_like(gray_image, dtype=float)

    for y in range(1, gray_image.shape[0] - 1):
        for x in range(1, gray_image.shape[1] - 1):
            gradient_x[y, x] = np.sum(prewitt_x * gray_image[y-1:y+2, x-1:x+2])
            gradient_y[y, x] = np.sum(prewitt_y * gray_image[y-1:y+2, x-1:x+2])

    gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2).astype(np.uint8)

    return gradient_magnitude

# Load an image using PIL
#change path here:
image = Image.open('/content/drive/MyDrive/OIP.jpeg')
image_array = np.array(image)
prewitt_result = prewitt_operator(image_array)
```

# Applying laplacian operator without any built-in functon:


```python


def laplacian_operator(image):
    gray_image = grayscale(image)

    laplacian_kernel = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])

    laplacian_result = np.zeros_like(gray_image, dtype=float)

    for y in range(1, gray_image.shape[0] - 1):
        for x in range(1, gray_image.shape[1] - 1):
            laplacian_result[y, x] = np.sum(laplacian_kernel * gray_image[y-1:y+2, x-1:x+2])

    laplacian_result = np.clip(laplacian_result, 0, 255).astype(np.uint8)

    return laplacian_result

# Load an image using PIL
#change path here:
image =  Image.open('/content/drive/MyDrive/OIP.jpeg')
image_array = np.array(image)
laplacian_result = laplacian_operator(image_array)
```

# Applying robert operator without any built-in functon:


```python


def robert_operator(image):
    gray_image = grayscale(image)

    robert_x = np.array([[1, 0], [0, -1]])
    robert_y = np.array([[0, -1], [1, 0]])

    gradient_x = np.zeros_like(gray_image, dtype=float)
    gradient_y = np.zeros_like(gray_image, dtype=float)

    for y in range(1, gray_image.shape[0] - 1):
        for x in range(1, gray_image.shape[1] - 1):
            gradient_x[y, x] = np.sum(robert_x * gray_image[y-1:y+1, x-1:x+1])
            gradient_y[y, x] = np.sum(robert_y * gray_image[y-1:y+1, x-1:x+1])

    gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2).astype(np.uint8)

    return gradient_magnitude

# Load an image using PIL
#change path here:
image =  Image.open('/content/drive/MyDrive/OIP.jpeg')
image_array = np.array(image)
robert_result = robert_operator(image_array)
```

# Function to plot images:


```python
import matplotlib.pyplot as plt

def plot_image(image, title):
    plt.figure()
    plt.imshow(image,cmap='gray', vmin=0, vmax=255)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Assuming you have the result images like "sobel_result", "robert_result", "laplacian_result", "prewitt_result"
plot_image(sobel_result, 'Sobel Operator')
plot_image(robert_result, 'Robert Operator')
plot_image(laplacian_result, 'Laplacian Operator')
plot_image(prewitt_result, 'Prewitt Operator')
```
